# now start the algorithm

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#version 1: calculate distance by points(rows)
class knn_version1():
    def __init__(self):
        pass
    
    def __init__(self, k):
        self.k = k
        
    def fit(self, xtrain, ytrain): # can also call it train
        self.xtrain = xtrain
        self.ytrain = ytrain
        
    def predict(self, xtest):
        
        trainLen = len(self.xtrain)
        testLen = len(xtest)
        
        #pred = []# got index over range error
        pred = np.zeros(xtest.shape[0])
        for i in range(testLen):
            
            distances = []
            
            for j in range(trainLen): # should be testLen not trainLen
                
                dist = self.calculatePointDistance(xtest[i], xtrain[j])
                #print(i, j, dist, sep = "\t")
                distances.append(dist)

            sortedIndex = np.argsort(distances)
            
            # get the k smallest distances and its class/label
            classCount = {}
            for m in range(k): # cant use i anymore
                index = sortedIndex[m]
                label = ytrain[index]
                classCount[label] = classCount.get(label, 0) + 1
            
            max = 0
             
            for key, v in classCount.items():
                if v > max:
                    max = v
                    #print(i , j , key, sep = "\t")
                    pred[i] = key
                    
        return pred
    
    def calculatePointDistance(self, A, B):
        ans = (((A-B)**2).sum())**0.5
        return ans
```

```python
#ccuracyScores = []
maxScore = 0
finalK = 0
for k in range(1,10): # k cant be 0
    model = knn_version1(k)
    #model.train(xtrain, ytrain)
    model.fit(xtrain, ytrain)
    pred = model.predict(xtest)
    accuracyScore = accuracy_score(ytest, pred)
    #ccuracyScores.append(accuracyScore)
    print("k = " + str(k) + " accuracyScore is " + str(accuracyScore))
    if accuracyScore > maxScore:
        maxScore = accuracyScore
        finalK = k
print(" The best k is " + str(finalK))
```

```python
#version2: calculate the distances by row-matrix(one row in xtest vs the entire xtrain)

class knn_version2():
    def __init__(self):
        pass
    
    def __init__(self, k):
        self.k = k
        
    def train(self, xtrain, ytrain):
        self.xtrain = xtrain
        self.ytrain = ytrain
        
    def predict(self, xtest):
        
        testLen = len(xtest)
        
        #pred = []# got index over range error
        pred = np.zeros(xtest.shape[0])
        for i in range(testLen):
                        
            distances = self.calculateDistances(xtest[i], xtrain)
            sortedIndex = np.argsort(distances)
            
            # get the k smallest distances and its class/label
            classCount = {}
            for m in range(k): # cant use i anymore
                index = sortedIndex[m]
                label = ytrain[index]
                classCount[label] = classCount.get(label, 0) + 1
            
            max = 0
             
            for key, v in classCount.items():
                if v > max:
                    max = v
                    #print(i , j , key, sep = "\t")
                    pred[i] = key
                    
        return pred
    
    # the only difference would be this function call
    # we calculate the distances not by point to point
    # but by row to entire xtrain
    # result is an collection of distances, with all the distance between this 
    # test point and all the points in the xtrain
    def calculateDistances(self, xtestRow, xtrain):
        xtrainLen = len(xtrain)
        distances = []
        
        xtestMatrix = np.tile(xtestRow, (xtrainLen, 1)) # repeat the xtest row in (trainLen, 1) shape
        diff = xtestMatrix - xtrain
        #result = ((diff ** 2).sum()) ** 0.5 # remember to use axis = 1
        result = ((diff ** 2).sum(axis = 1)) ** 0.5
        return result
        
```

```python
#ccuracyScores = []
maxScore = 0
finalK = 0
for k in range(1,10): # k cant be 0
    model = knn_version2(k)
    model.train(xtrain, ytrain)
    pred = model.predict(xtest)
    accuracyScore = accuracy_score(ytest, pred)
    #ccuracyScores.append(accuracyScore)
    print("k = " + str(k) + " accuracyScore is " + str(accuracyScore))
    if accuracyScore > maxScore:
        maxScore = accuracyScore
        finalK = k
    #print("finalK is " +  str(finalK))
print(" The best k is " + str(finalK))
```

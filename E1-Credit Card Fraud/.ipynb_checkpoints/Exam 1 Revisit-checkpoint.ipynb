# Visualize data

```python
len = train_origin.shape[1]

def draw_histograms(df, variables, n_rows, n_cols):
    fig=plt.figure(figsize=(10,80))
    for i, var_name in enumerate(variables):
        ax=fig.add_subplot(n_rows,n_cols,i+1)
        df[var_name].hist(bins=20,ax=ax)
        ax.set_title(var_name+" Distribution")
    fig.tight_layout()  # Improves appearance a bit.
    plt.show()
    
draw_histograms(train_origin, train_origin.columns, len/2, 2)
```

# Feture Selection
# Using Cross validation with Naive Bayes to find the most unrelated columns

```python
y = train_origin["Class"]
X = train_origin.drop("Class", axis = 1)
y.head()
```

```python
X.head()
```

```python
from sklearn.naive_bayes import BernoulliNB
def NB_classify(drop_var, X, y):
    clf = BernoulliNB()
    scores = cross_val_score(clf, X.drop(drop_var, axis = 1), 
                             y, cv = 4, scoring = "f1")
    #print(np.mean(scores))
    return np.mean(scores)
```

```python
scores = dict()
for col in X.columns:
    #print(col)
    score = NB_classify(col, X, y)
    scores[col] = score
```

```python
print(*sorted(scores.items(), key = lambda x:x[1]), sep = "\n")
```

# use sklearn class method selectKBest to do feature filtering

```python
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(f_classif, k = 29).fit(X, y)
features = selector.transform(X)

mask = selector.get_support()
newFeatures = []
abandanedFeatures = []
for bool, feature in zip(mask, X.columns):
    if  bool:  
        newFeatures.append(feature)
    else:
        abandanedFeatures.append(feature)

X_new = pd.DataFrame(features, columns = newFeatures )
abandanedFeatures
```

# drop columns most unrelated

```python
abandonedColumns = ["V15", "V18", "V5"]
selectedX = X.drop(abandonedColumns, axis = 1)
selectedXtest = test_origin.drop(abandonedColumns, axis = 1)
```

# training with Naive Bayes model

```python
model_nb = BernoulliNB()
model_nb.fit(selectedX, y)

pred_nb = model_nb.predict(selectedXtest)
```

```python
submission_nb = pd.DataFrame({"Id": test_origin.Id, "Class": pred_nb})
submission_nb.to_csv("./revisit_output/submission_nb_revisit_Jie_Ling.csv", columns=["Id", "Class"], header = True, index = False)
```

# Data sampling for unbalanced data

```python
def under_sampling(data, ratio):
    #print("ratio " + str(ratio))
    count_bad = len(train_origin[train_origin.Class == 1])
    index_bad = np.array(train_origin[train_origin.Class == 1].index)
    
    index_good = train_origin[train_origin.Class == 0].index
    index_good_random = np.random.choice(index_good, count_bad * ratio, replace= False)
    
    sampled_index = np.concatenate([index_bad, index_good_random])
    sampled_data  = train_origin.iloc[sampled_index, :]
    
    X = sampled_data.iloc[:, sampled_data.columns != "Class" ]
    y = sampled_data.Class
    return X, y
```

```python
X_train_undersampled, y_train_undersampled = under_sampling(train_origin, 250)
model_nb_undersampled = BernoulliNB()
model_nb_undersampled.fit(X_train_undersampled.drop(["V5", "V18", "V1"], axis = 1), y_train_undersampled)
```

```python
pred_nb_undersampled = model_nb_undersampled.predict(test_origin.drop(["V5", "V18", "V1"], axis = 1))
```

```python
submission_nb_undersampled = pd.DataFrame({"Id": test_origin.Id, "Class": pred_nb_undersampled})
submission_nb_undersampled.to_csv("./revisit_output/submission_nb_revisit_undersampled.csv", columns=["Id", "Class"], header = True, index = False)
```

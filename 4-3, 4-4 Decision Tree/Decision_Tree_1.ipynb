```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)

N = 500
D = 2
X = np.random.randn(N, D)  #  创建n*d高丝分布
print(type(X))
print(X.shape)
```

```python
Y = np.array([0] * (N//2) + [1] * (N//2))
# how ridiculous is this way of initialization!!!
print(Y)
print(Y.shape)
```

```python
plt.scatter(X[:,0], X[:,1], s = 100, c = Y, alpha = 0.5)
```

```python
delta = 1.5
np.array([delta, delta])
```

```python
X
```

```python
X[:N//2] += np.array([delta, delta])
```

```python
X[N//2:] += np.array([-delta, -delta])
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1], s=100, c=Y, alpha = 0.5 )
```

```python
def plot_decision_boundary(X, model):
    h = .02
    
    x_min, x_max = X[:, 0].min() -1, X[:,0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 0].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)
    
    plt.contour(xx, yy, z, cmap=plt.cm.Paired)
```

# start the model

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X, Y)
print(" score for basic tree is ", model.score(X, Y))
```

```python
plt.scatter(X[:,0], X[:, 1], s=100, c=Y, alpha=0.5)
plot_decision_boundary(X, model)
plt.show()
```

model with depth as 3

```python
model_depth_3 = DecisionTreeClassifier(criterion="entropy", max_depth = 3)
model_depth_3.fit(X, Y)
print(" score for basic tree ", model_depth_3.score(X, Y))
```

```python
plt.scatter(X[:, 0], X[:, 1], s = 100, c = Y, alpha = 0.5)
plot_decision_boundary(X, model_depth_3)
plt.show()
```

model of depth 5

```python
model_depth_5 = DecisionTreeClassifier(criterion= "entropy", max_depth = 5)
model_depth_5.fit(X, Y)
print(" score of basic tree ", model_depth_5.score(X, Y))
```

```python
plt.scatter(X[:,0], X[:, 1], s=100, c=Y, alpha = 0.5)
plot_decision_boundary(X, model_depth_5)
plt.show()
```
